```python
import sys
sys.setrecursionlimit(10000)


def dfs(graph, visited, now, prev, colors, has_cycle):

    if has_cycle == False:

        visited[now] = True
        colors[now] = 1

        if len(graph[now]) == 1:
            colors[now] = 2
            

        for neig in graph[now]:
            if neig != prev and not visited[neig]:
                has_cycle = dfs(graph, visited, neig, now, colors, has_cycle)
            elif colors[neig] == 1 and neig != prev and not has_cycle:
                has_cycle = True
                return has_cycle
                
    return False
        
    

with open("input_35.txt", "r") as file:

    N = int(file.readline())

    graph = [[0]]*(N + 1)
    visited = [False] * (N + 1)

    for i in range(N + 1):
        if i == 0:
            pass
        else:
            inp = list(map( int, file.readline().split(sep = ' ')))

            for j in range(len(inp)):

                if inp[j] == 1:
                    if 0 in graph[i]:
                        graph[i].pop()
                    graph[i] = graph[i] + [j+1]

    colors = [0] * (N + 1)
    has_cycle = False
    prev = 0

    for i in range(1, N + 1):
            if not visited[i]:
                has_cycle = dfs(graph, visited, i, prev, colors, has_cycle)
            if has_cycle == True:
                break


    if has_cycle == False:
        print('NO')
    else:
        print('YES')
        ans = []
        cnt = 0
        for i in range(len(colors)):
            if colors[i] == 1:
                cnt += 1
                ans.append(i)
        print(cnt)
        print(*ans)
```
